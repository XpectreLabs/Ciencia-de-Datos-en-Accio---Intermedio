# Ejercicio 2.1: Visualización de Ventas por Región (Ejemplo)

En este ejercicio, trabajarás con un dataset simple llamado `simple_sales_data.csv` que contiene información sobre ventas en diferentes regiones. Crearás un gráfico de barras para visualizar las ventas totales por región, utilizando pandas para agrupar los datos y Matplotlib para la visualización.

Tareas:
1. Cargar el dataset `simple_sales_data.csv`.
2. Agrupar las ventas por región y calcular las ventas totales.
3. Crear un gráfico de barras para mostrar las ventas totales por región.

```python
# Paso 1: Importar las librerías necesarias
# -------------------------------------------
# Asegúrate de tener instaladas las librerías pandas y matplotlib.

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Paso 2: Cargar el dataset desde un archivo local
# -------------------------------------------
# Usa la opción "Archivos" en Google Colab para cargar el archivo `simple_sales_data.csv`.

df = pd.read_csv("simple_sales_data.csv")
```

```python
# Paso 3: Agrupar las ventas por región y calcular las ventas totales
# -------------------------------------------
# Usa `groupby` para agrupar los datos por la columna `Region` y sumar las ventas.

sales_by_region = df.groupby("Region")["Sales"].sum()
```

```python
# Paso 4: Crear un gráfico de barras
# -------------------------------------------
# Utiliza Matplotlib para visualizar las ventas totales por región en un gráfico de barras.

sales_by_region.plot(kind="bar", color="skyblue")
plt.title("Ventas Totales por Región")  # Título del gráfico
plt.ylabel("Ventas Totales")           # Etiqueta del eje Y
plt.xlabel("Región")                   # Etiqueta del eje X
plt.show()
```

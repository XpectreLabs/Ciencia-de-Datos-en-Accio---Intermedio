# Ejercicio 2.2: Boxplot de Calificaciones (Ejemplo)

En este ejercicio, trabajarás con un dataset mediano llamado `medium_students_data.csv`, que contiene las calificaciones de estudiantes en tres materias: Matemáticas, Ciencias y Inglés. Crearás un boxplot para visualizar la distribución de las calificaciones de cada materia y analizar posibles valores atípicos y tendencias.

Tareas:
1. Cargar el dataset `medium_students_data.csv`.
2. Seleccionar las columnas de calificaciones de las tres materias.
3. Crear un boxplot para mostrar la distribución de calificaciones por materia.

```python
# Paso 1: Importar las librerías necesarias
# -------------------------------------------
# Asegúrate de tener instaladas las librerías pandas y matplotlib.

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Paso 2: Cargar el dataset desde un archivo local
# -------------------------------------------
# Usa la opción "Archivos" en Google Colab para cargar el archivo `medium_students_data.csv`.

df = pd.read_csv("medium_students_data.csv")
```

```python
# Paso 3: Seleccionar las columnas de calificaciones
# -------------------------------------------
# Filtra las columnas `Math`, `Science` y `English` para trabajar únicamente con las calificaciones.

grades = df[["Math", "Science", "English"]]
```

```python
# Paso 4: Crear un boxplot para visualizar la distribución
# -------------------------------------------
# Utiliza la función `boxplot` de pandas para generar un gráfico que muestre la distribución de calificaciones por materia.

grades.boxplot()
plt.title("Distribución de Calificaciones por Materia")  # Título del gráfico
plt.ylabel("Calificaciones")                            # Etiqueta del eje Y
plt.show()
```

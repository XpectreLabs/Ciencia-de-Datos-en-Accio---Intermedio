# Ejercicio 3.1: Relación entre Variables (Ejemplo)

En este ejercicio, trabajarás con el dataset `medium_students_data.csv`, que contiene las calificaciones de estudiantes en Matemáticas y Ciencias. Crearás un scatter plot para visualizar la relación entre las calificaciones en Matemáticas y Ciencias y analizar cómo están relacionadas estas dos materias.

Tareas:
1. Cargar el dataset `medium_students_data.csv`.
2. Crear un scatter plot para visualizar la relación entre calificaciones en Matemáticas y Ciencias.


```python
# Paso 1: Importar las librerías necesarias
# -------------------------------------------
# Asegúrate de tener instaladas las librerías pandas, seaborn y matplotlib.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Paso 2: Cargar el dataset desde un archivo local
# -------------------------------------------
# Usa la opción "Archivos" en Google Colab para cargar el archivo `medium_students_data.csv`.

df = pd.read_csv("medium_students_data.csv")
```

```python
# Paso 3: Crear un scatter plot para la relación entre Matemáticas y Ciencias
# -------------------------------------------
# Utiliza la función `sns.scatterplot()` para generar un gráfico de dispersión que muestre la relación entre las calificaciones en Matemáticas y Ciencias.

sns.scatterplot(x="Math", y="Science", data=df)
plt.title("Relación entre Matemáticas y Ciencias")  # Título del gráfico
plt.xlabel("Calificaciones en Matemáticas")         # Etiqueta del eje X
plt.ylabel("Calificaciones en Ciencias")            # Etiqueta del eje Y
plt.show()
```

# Ejercicio 3.2: Heatmap de Correlaciones del Titanic (Ejemplo)

En este ejercicio, trabajarás con el dataset Titanic para analizar las correlaciones entre las variables numéricas. Crearás un heatmap utilizando `seaborn` para visualizar cómo están relacionadas estas variables.

Tareas:
1. Cargar el dataset `titanic.csv`.
2. Crear un heatmap para visualizar las correlaciones entre las variables numéricas.


```python
# Paso 1: Importar las librerías necesarias
# -------------------------------------------
# Asegúrate de tener instaladas las librerías pandas, seaborn y matplotlib.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Paso 2: Cargar el dataset desde un archivo local
# -------------------------------------------
# Usa la opción "Archivos" en Google Colab para cargar el archivo `titanic.csv`.

df = pd.read_csv("titanic.csv")
```

```python
# Paso 3: Verifica las primeras filas del DataFrame para identificar el problema
# -------------------------------------------
print(df.head())

# Paso 4: Identificar columnas no numéricas
# -------------------------------------------
non_numeric_cols = df.select_dtypes(include='object').columns
print(f'Columnas no numéricas: {non_numeric_cols}')
```

```python

# Paso 5: Elimina las columnas no numéricas del DataFrame
# -------------------------------------------
df_numeric = df.drop(columns=non_numeric_cols)

# Paso 6: Verifica el DataFrame después de la limpieza
# -------------------------------------------
print(df_numeric.head())
```

```python
# Paso 7: Realiza el heatmap
# -------------------------------------------
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Mapa de Correlaciones del Dataset Titanic")  # Título del heatmap
plt.show()
```

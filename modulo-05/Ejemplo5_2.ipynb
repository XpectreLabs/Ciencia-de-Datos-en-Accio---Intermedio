 # Ejemplo 5.2: Reducción de Dimensionalidad con PCA (Análisis de Componentes Principales)

Descripción:
En este ejercicio, vamos a aplicar PCA (Principal Component Analysis) para reducir la dimensionalidad de un conjunto de datos.
Usaremos el dataset de Iris para aplicar PCA y reduciremos la dimensionalidad de 4 características a 2 componentes principales.

Tareas:
1. Cargar el dataset de Iris.
2. Estándarizar los datos para que todas las características tengan la misma escala.
3. Aplicar PCA para reducir la dimensionalidad de 4 a 2 componentes principales.
4. Visualizar los datos proyectados en los dos primeros componentes principales.


```python
# Paso 1: Importar las librerías necesarias
# -------------------------------------------
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Paso 2: Cargar el dataset de Iris
# -------------------------------------------
# Cargar el dataset de Iris, que tiene 150 muestras y 4 características.
iris = load_iris()
X = iris.data  # Características
y = iris.target  # Etiquetas (especies)
```

```python
# Paso 3: Estándarizar los datos
# -------------------------------------------
# Estandarizamos las características para que todas tengan la misma escala.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Paso 4: Aplicar PCA
# -------------------------------------------
# Reducimos la dimensionalidad de 4 a 2 componentes principales.
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Paso 5: Visualizar los datos proyectados
# -------------------------------------------
# Creamos un gráfico de dispersión usando los dos primeros componentes principales.
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.title("Proyección del Dataset de Iris en 2 Componentes Principales")
plt.xlabel("Primer Componente Principal")
plt.ylabel("Segundo Componente Principal")
plt.colorbar(label="Especie")
plt.show()
```

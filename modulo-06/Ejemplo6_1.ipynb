# Análisis de Ventas y Crecimiento por Región**  
**Objetivo:** Visualizar las ventas y el crecimiento porcentual por región para identificar tendencias y áreas de oportunidad.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# **Paso 1: Creación del Dataset**
# Creamos un DataFrame con datos ficticios sobre ventas y crecimiento porcentual por región.
data = pd.DataFrame({
    'Region': ['Norte', 'Sur', 'Este', 'Oeste'],  # Regiones geográficas
    'Ventas': [25000, 15000, 20000, 10000],  # Ventas en USD por región
    'Crecimiento (%)': [5.2, 3.8, 4.5, 2.1]  # Crecimiento porcentual de ventas por región
})
```

```python
# **Paso 2: Visualización de Ventas por Región**
plt.figure(figsize=(8, 6))  # Definir el tamaño del gráfico
sns.barplot(x='Region', y='Ventas', data=data, palette='Blues')  # Gráfico de barras
plt.title('Ventas por Región', fontsize=14)  # Título del gráfico
plt.xlabel('Región', fontsize=12)  # Etiqueta del eje X
plt.ylabel('Ventas (USD)', fontsize=12)  # Etiqueta del eje Y
plt.show()  # Mostrar el gráfico
```

```python
# **Paso 3: Visualización del Crecimiento por Región**
plt.figure(figsize=(8, 6))  # Definir el tamaño del gráfico
sns.lineplot(x='Region', y='Crecimiento (%)', data=data, marker='o', color='green')  # Gráfico de líneas
plt.title('Crecimiento por Región', fontsize=14)  # Título del gráfico
plt.xlabel('Región', fontsize=12)  # Etiqueta del eje X
plt.ylabel('Crecimiento (%)', fontsize=12)  # Etiqueta del eje Y
plt.show()  # Mostrar el gráfico

# **Narrativa:**
# Ventas por región: El Norte lidera las ventas con $25,000 USD, seguido del Este con $20,000 USD.
# Crecimiento: Aunque el Oeste tiene las menores ventas, registra un crecimiento positivo del 2.1%.
# **Recomendaciones:**
# - Invertir en estrategias para aumentar las ventas en el Oeste.
# - Mantener el liderazgo en el Norte potenciando promociones.
```
